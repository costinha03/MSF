```python
#1a


import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols
import sympy as sy

t = np.arange(0,200,1) #definir o tempo

#carro v é constante e igual a 70 kmh
#cp v0 = 0 e a=2 m/s^2

x0 = 0
v0 = 0
v = 70/3.6
a = 2

xc = v * t
xcp = t**2

plt.xlabel("t(s)")
plt.ylabel("x(t)(m)")

plt.plot(t,xc, label="Carro A")
plt.plot(t,xcp, label="Polícia")
plt.ylim([-1, 1000])
plt.xlim([-1, 50])
plt.legend()

#1b


```

```python
for t_test in range(0,50000):
    t_test /= 1000

    if 19.44 * t_test < 0.5 * 2 * t_test**2:
        print("Encontram-se ao final de ", t_test, "segundos")
        break


print("Percorreu", 19.44*t_test, "metros")

t, v0, a  = symbols('t, v0, a')

xc = v0 * t
xcp = 0.5 * a * t**2

print(xc)
print(xcp)

xc = xc.subs([(v0, (70/3.6))])
xcp = xcp.subs([(a, 2)])

print(xc)
print(xcp)

tint = sy.nsolve(xc-xcp,t,20)

print("\n Intercetaram-se ao fim de {} s".format(tint))

xfin = xc.subs([(t, tint)])
print("\n Percorreram {} metros antes de se intercetarem".format(xfin))
```
