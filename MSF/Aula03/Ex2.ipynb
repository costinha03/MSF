```python
#b
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy

tempo = np.linspace(0, 4)

vT, g, t = sy.Symbol("vT, g, t")


d = sy.Derivative((vT**2)/g * sy.log(sy.cosh(g*t/vT)), t)
d_simp = sy.simplify(d)

print("v(t) = ", d_simp)

plt.xlabel("Tempo (s)")
plt.ylabel("Velocidade (m/s)")
plt.plot(tempo, 6.80*np.tanh(9.8*tempo/6.80))
```
