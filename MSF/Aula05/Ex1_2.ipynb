```python
import numpy as np
import matplotlib.pyplot as mp
import sympy as sy

def norma(x):
    result = 0
    for i in range(len(x)):
        result += x[i]**2
    return result ** 0.5

print(norma(np.array([3,4])))
```

```python
def unitary(x):
    result = []
    for i in range(len(x)):
        result.append(x[i]/norma(x))
    return result

print(unitary([3,4]))
```

```python
def produtoesc(x,y):
    assert len(x) == len(y)
    result = 0
    for i in range (len(x)):
        result += x[i] * y[i]
    return result

print(produtoesc([1,2],[-2,3]))

def angle(x,y):
    res1 = np.dot(x,y)
    res2 = norma(x) * norma(y) 
    result = np.arccos(res1 / res2)
    result = np.degrees(result)
    return result

print(angle(np.array([1,2]), np.array([3,4])))
```
