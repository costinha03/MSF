```python
import matplotlib.pyplot as plt

t = [1, 2, 3, 4]
r = [(2*t[i], t[i]**2) for i in range(len(t))]

plt.plot([0, r[0][0]], [0, r[0][1]], 'bo-', label='t=1')
plt.plot([0, r[1][0]], [0, r[1][1]], 'go-', label='t=2')
plt.plot([0, r[2][0]], [0, r[2][1]], 'ro-', label='t=3')
plt.plot([0, r[3][0]], [0, r[3][1]], 'yo-', label='t=4')

plt.legend()
plt.show()
```
