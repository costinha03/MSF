```python
import matplotlib.pyplot as mpl
import numpy as np;





def euler(ti,tf,dt):
    g = 9.81
    n = int((tf-ti)/dt)
    vx = np.zeros(n+1)
    vy = np.zeros(n+1)
    t = np.zeros(n+1)
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    v = np.zeros(n+1)
    ax = np.zeros(n+1)
    ay = np.zeros(n+1)
    ang = np.deg2rad(10)
    v0 = 100 / 3.6
    vy[0] = np.sin(ang) * v0
    vx[0] = np.sin(ang) * v0
    x[0] = 0
    y[0] = 0
    ay[0] = -g
    ax[0] = 0
    v[0] = v0
    vT = 100 / 3.6
    D = g / vT**2
    g = 9.8

    for i in range(n):
        v[i] = np.sqrt(vy[i]**2 + vx[i]**2)
        ay[i] = -g - D*vy[i]*v[i]
        ax[i] =  - D*vx[i]*v[i]
        vy[i+1] = vy[i] + ay[i]*dt
        vx[i+1] = vx[i] + ax[i]*dt
        y[i+1] = y[i] + vy[i]*dt
        x[i+1] = x[i] + vx[i]*dt
        t[i+1] = t[i] + dt
        
    return t, vx, vy, x, y

t, vx, vy, x, y = euler(0,1,0.1)

g = 9.81
xExato = x[0] + vx[0]*t
yExato = y[0] + vy[0]*t - 1/2*g*t**2
mpl.plot(xExato,yExato, label="exato" )
mpl.plot(x,y, label = "Euler com resistÃªcia dt = 0.1")
mpl.legend()

i = np.where(vy < 0)[0][0]

print(y[i])
```
