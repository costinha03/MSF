```python
import numpy as np
import matplotlib.pyplot as plt


def euler_ciclista(t0,tf,v0x,x0,w,dt,Cres,Pot):  # w é angulo que pode aparecer em alguns exs

    g = 9.8

    n = int((tf-t0)/dt+0.1)
    t = np.zeros(n+1)
    x = np.zeros(n+1)
    vx = np.zeros(n+1)
    ax = np.zeros(n+1)

    x[0]  = x0
    vx[0] = v0x
    t[0] = t0

    p_ar = 1.225  #densidade do ar
    coef_u = 0.004
    m = 75 #massa ciclista-bicicleta
    cRes = 0.9 #Coeficiente da res ar
    A = 0.3 # Área frontal
    w = np.deg2rad(w)

    for i in range(n):
        ax[i] =(Pot / (m * vx[i])) -( g * np.sin(w)) - ((Cres /(2*m))*A*p_ar*vx[i]*vx[i])-(coef_u*g) #slide38
        vx[i+1] = vx[i]+ax[i]*dt
        x[i+1] = x[i] + vx[i] * dt
        t[i+1] = t[i]+dt
    return t,x,vx,ax,n

t,x,vx,ax,n = euler_ciclista(0,600,1,0,5,0.001,0.9,0.4*735)



i = np.where(ax<=0.0)[0][0]
vt = vx[i]
print("Velocidade terminal é {} m/s".format(round(vt, 3)))
i2 = np.where(vx>=0.9*vt)[0][0]
print("Atingiu 90% da velocdade terminal no instante t = {} s".format(round(t[i2],3)))
i3 = np.where(x>=2000)[0][0]
print("Demorou {} s a percorrer 2km".format(round(t[i3],3)))

plt.plot(t, vx, color="red")
plt.xlabel("Tempo (s)")
plt.ylabel("Velocidade (m/s)")
plt.grid()
# plt.show()

```
